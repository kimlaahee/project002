```python
# 출산율 예측
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt

# 한글 폰트
font_path='NanumGothic.ttf' 
font_name=font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name) 

data2 = {
    'year': [2014,2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'pri_edu': [24.2, 24.4, 25.6, 27.2, 29.1, 32.1, 30.2, 36.7, 41.0, 43.4]  
}
df2 = pd.DataFrame(data2)


x = df2[['year']] 
y = df2['pri_edu']  

model2 = LinearRegression()
model2.fit(x, y)  #학습 후 
 

f_years = pd.DataFrame({'year': [2024, 2025, 2026, 2027, 2028]})


p_birth_edu = model2.predict(f_years) # 예측 


f_df2 = pd.DataFrame({'year': f_years['year'],\
                          'pri_edu': p_birth_edu})

# 출산율

data2 = {
    'year': [2014,2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'bir_rate': [43.9, 44.4, 41.2, 36.2, 33.4, 30.8, 27.6, 26.3, 25.5, 23.5]  
}
df3 = pd.DataFrame(data2)


x = df3[['year']] 
y = df3['bir_rate']  

model3 = LinearRegression()
model3.fit(x, y)  #학습 후  

f_years = pd.DataFrame({'year': [2024, 2025, 2026, 2027, 2028]})
p_birth_edu = model3.predict(f_years) # 예측 


f_df3 = pd.DataFrame({'year': f_years['year'],\
                          'bir_rate': p_birth_edu})

fig = plt.figure(figsize = (15, 15))
axes2 = fig.add_subplot(111)
# axes1.plot(df2['year'], df2['pri_edu'], marker='*',label='사교육 비용(2014~2023)', color='green')
# axes1.plot(f_df2['year'], f_df2['pri_edu'], marker='*',label='미래 사교육 비용 추이', color='red')
# axes1.legend(fontsize = 15)
# axes1.set_ylabel('사교육 비용(만 원)', fontsize = 20, fontweight = 'bold')

# axes2 = axes1.twinx()
axes2.plot(df3['year'], df3['bir_rate'], marker='o',label='출산율 (2014~2023)', color='blue')
axes2.plot(f_df3['year'], f_df3['bir_rate'], marker='o',label='미래 출산율 추이', color='red')
axes2.legend(fontsize = 15)
axes2.set_ylabel('출산율(만 명)', fontsize = 20, fontweight = 'bold')

plt.rc('font', family = 'Malgun Gothic', size = '14') # 한글 처리
plt.title('향후 5년 출산율 예측치',fontweight = 'bold', pad=20)
plt.grid(True)
# axes1.legend(loc='lower left', bbox_to_anchor=(0,1.0))
axes2.legend(loc='lower right', bbox_to_anchor=(1.0,1.0))
plt.xticks(range(2014,2029,1))

plt.savefig('향후 5년 출산율 예측치.jpg')
plt.show()
```
