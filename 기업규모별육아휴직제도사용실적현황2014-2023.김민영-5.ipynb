```python
# 육아휴직제도사용실적현황.csv 파일 읽기
import pandas as pd
import numpy as np

df=pd.read_csv('육아휴직제도사용실적현황.csv', low_memory=False, encoding='cp949')
df
```

```python
### Dataframe Cleansing
## 빈도(개)와 비율(%) 중에 비율(%)만 추출

columns2=['사업체특성별(1)','사업체특성별(2)']

for i in range(6):
    a=2017.1
    b=2017.3
    a+=i
    columns2.append(str(a))
    b+=i
    columns2.append(str(b))           
    
print(columns2) # 확인용 

df2=df[columns2]
df2
```

```python
# [csv파일 CLEANSING 및 FIXING]
# 컬럼명 변경

columns3=['사업체특성별(1)','사업체특성별(2)']

for i in range(6):
    a=2017
    b=2017
    a+=i
    columns3.append(str(a)+' 있다')
    b+=i
    columns3.append(str(b)+' 없다')           
    
print(columns3) # 확인용 

# 컬럼명 변경
df2.columns = columns3
df2

# 인덱스: '사업체특성별(1)','사업체특성별(2)'를 인덱스로 변경
df2.set_index(['사업체특성별(1)','사업체특성별(2)'], inplace=True)
df2
    
```

```python
# 원본 데이터프레임에서 첫 3행 삭제
df3 = df2.drop(df2.index[:4],axis=0)
df3 = df3.astype(float)
df3
```

```python
# # 인덱스가 '종사자규모별'인 행들 선택
size = df3.xs('종사자규모별',level=0)
size
```

```python
size2=size.T
size2

# 인덱스명, 컬럼명 변경
size2.index.name='사용실적현황'
size2.columns.name='종사자규모'
size2
```

```python
## 위의 과정을 함수로 만들어서 함수 호출로 변경

# 연도별 데이터 프레임을 생성
def size59_df(df, year):
    size = df[[f'{year} 있다', f'{year} 없다']]
    size.index = [year]
    size.columns = ['있다', '없다']
    return size

# 각 연도에 대해 데이터 프레임 생성
years = [2017,2018, 2019, 2020,2021, 2022]
size_dfs = [size59_df(size_df4, year) for year in years]

# size_dfs 는 리스트
# size_dfs

# [       있다    없다
#  2017  2.8  97.2,
#         있다    없다
#  2018  2.5  97.5,
#         있다    없다
#  2019  3.3  96.7,
#         있다    없다
#  2020  3.8  96.2,
#         있다    없다
#  2021  9.6  90.4,
#         있다    없다
#  2022  6.1  93.9]

# 데이터 프레임 결합
size_59 = pd.concat(size_dfs)

# 결과 출력
size_59.index.name='연도'
size_59.columns.name='사용현황'
size_59
```
