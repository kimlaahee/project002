```python
import pandas as pd
import numpy as np

# 엑셀파일==> Dataframe으로 변환
df=pd.DataFrame(pd.read_excel('출산 및 육아휴직급여 수급자 현황 2014_2023.xlsx'))
df
```

```python
# 위의 df에서 10~12행 삭제
df2 = df.drop(df.index[10:13],axis=0)
df2
```

```python
# df2에서 2행을 컬럼으로 지정 예정
# 컬럼 변경 시 인덱스명이 될 부분을 미리 변환
df2.at[1,'통계표명:']= '분류'
df2
```

```python
# 컬럼이 될 NaN값을 임의의 문자로 변경
df2.at[1,'출산 및 육아휴직급여 수급자 현황']= '없어'
df2
```

```python
# 1행을 컬럼으로 지정
df3 = df2.rename(columns=df2.iloc[1])
df3
```

```python
# 필요없는 행(0,1행) 삭제
df4 = df3.drop(df3.index[0:4],axis=0) ########
df4

# df4.info()
```

```python
#  인덱스 변경==> 0,1,2,3....
df4.reset_index(drop = False, inplace = True)
df4
```

```python
# 필요없는 행(1,2행) 삭제
df4 = df4.drop(df4.index[1:3],axis=0)
df4
```

```python
#  인덱스 변경==> 0,1,2,3....
df4.reset_index(drop = False, inplace = True)
df4
```

```python
# 필요없는 행(2,3행) 삭제
df4 = df4.drop(df4.index[2:],axis=0)
df4
```

```python
# 인덱스 값 변경(==> '분류'열을 인덱스로 변경)
df4.set_index(['분류'],inplace=True)
df4
```

```python
# 필요없는 열 삭제: level_0, index, NaN
df4=df4.drop(['level_0','index','없어'], axis=1)
df4
```

```python
# 데이터 타입 확인
df4.dtypes
```

```python
# 데이터 타입 변경

def clean_fix(column):
    return column.str.replace(',', '').astype('int64')

for year in range(2014, 2024):
    df4[(str(year))] = clean_fix(df4[(str(year))])
    
df4
```

```python
df4=df4.T
df4
```

```python
df4.index.name='연도'
df4.columns.name=''
df4
```

```python
df4.info()
```

```python
df4.columns=['육아휴직급여 수급자 수(단위:명)','육아휴직급여 지원금액(단위:만원)']
df4=df4.reset_index()
df4
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rc('font',family='Malgun Gothic')
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 10

# 2. 그래프 그리기
fig, ax1 = plt.subplots()

# 꺾은선그래프 
ax1.plot(df4['연도'], df4['육아휴직급여 수급자 수(단위:명)'], '-s', color='green',\
         markersize=7, linewidth=5, alpha=0.7, label='육아휴직급여\n수급자 수(단위:명)')
ax1.set_ylim( 70000, 180000)
ax1.set_xlabel('연도')
ax1.set_ylabel('육아휴직급여 수급자 수(단위:명)')
ax1.tick_params(axis='both', direction='in')
ax1.set_title('육아휴직급여 수급자 현황 2014~2023')
# plt.grid(True)

# 막대 그래프
ax2 = ax1.twinx()
ax2.bar(df4['연도'], df4['육아휴직급여 지원금액(단위:만원)'], color='deeppink',\
        label='육아휴직급여\n지원금액(금액: 만원)', alpha=0.7, width=0.7)
ax2.set_ylim(300000, 2000000)
ax2.set_ylabel(r'육아휴직급여 지원금액(단위:만원)')
ax2.tick_params(axis='y', direction='in')

# 꺾은선 그래프 앞으로 보이게 함
ax1.set_zorder(ax2.get_zorder() + 10)
ax1.patch.set_visible(False)

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# # 각 데이터 포인트에 값 레이블 추가
for year, num in df4.iloc[:,[0,1]].values:
    ax1.annotate(num,(year,num),xytext=(3,10),textcoords='offset points', ha='center', fontsize=11 )

plt.savefig('육아휴직급여 수급자 현황 2014~2023.png')
plt.savefig('육아휴직급여 수급자 현황 2014~2023.jpg')
plt.show()
```
