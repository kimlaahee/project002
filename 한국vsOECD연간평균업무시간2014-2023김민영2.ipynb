```python
import pandas as pd
import numpy as np

df2=pd.read_csv('OECD.Average annual hours actually worked per worker.csv', low_memory=False)
df2
```

```python
df3=df2[['Reference area', 'TIME_PERIOD', 'OBS_VALUE']]
df3
```

```python
value2=(df3['Reference area']=='Korea')
df_kor=df3[value2]

df_kor
```

```python
value3=(df3['Reference area']=='OECD')
df_oecd2=df3[value3]

df_oecd2
```

```python
df_kor2=df_kor[['TIME_PERIOD','OBS_VALUE']]
# df_kor2

df_kor2=df_kor2.set_index('TIME_PERIOD')
df_kor2=df_kor2.sort_index()

df_kor2=df_kor2.reset_index()
df_kor2
```

```python
df_oecd2=df_oecd2[['TIME_PERIOD','OBS_VALUE']]

df_oecd2=df_oecd2.set_index('TIME_PERIOD')
df_oecd2=df_oecd2.sort_index()

df_oecd2=df_oecd2.reset_index()
df_oecd2
```

```python
line1=df_kor2
line2=df_oecd2

import matplotlib.pyplot as plt

plt.rc('font',family='Malgun Gothic')
plt.rc('font',size=15)

fig2,ax2=plt.subplots(figsize=(10,8))
ax2.plot(line1['TIME_PERIOD'],line1['OBS_VALUE'], '-s', color='deeppink', markersize=7, linewidth=4, alpha=0.7, \
        label='한국 평균 (단위: 인당, 시간)')
ax2.plot(line2['TIME_PERIOD'],line2['OBS_VALUE'], '-s', color='green', markersize=7, linewidth=4, alpha=0.7, \
        label='OECD 평균 (단위: 인당, 시간)')
ax2.set_title('한국 VS OECD 연간 평균 업무 시간 2014~2023(단위: 인당 시간)')
ax2.legend(loc='upper right')

# # 각 데이터 포인트에 값 레이블 추가
for year, num in line1.iloc[:,[0,1]].values:
    ax2.annotate(num,(year,num),xytext=(3,10),textcoords='offset points', ha='center', fontsize=11 )
    
for year, num in line2.iloc[:,[0,1]].values:
    ax2.annotate(num,(year,num),xytext=(3,10),textcoords='offset points', ha='center', fontsize=11 )

# 그래프의 테두리 컬러 및 그래프 안쪽 컬러 설정    
fig2.patch.set_facecolor('lightblue')    
ax2.set_facecolor('beige')    

plt.grid(True)
# fig2.show()

plt.savefig('한국 VS OECD 연간 평균 업무 시간 2014~2023.png')
plt.savefig('한국 VS OECD 연간 평균 업무 시간 2014~2023.jpg')
plt.show()
```
